```python
# Load the dataset (skip first 4 rows as required by World Bank format)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv", skiprows=4)

# Display first few rows
df.head()
```

```python
# Identify year columns dynamically (years are numeric)
year_columns = [col for col in df.columns if col.isdigit()]

# Get the latest available year
latest_year = max(year_columns)
print("Latest Year Available:", latest_year)
```

```python
# Select only 'Country Name' and the latest available year
df_selected = df[['Country Name', latest_year]].dropna()

# Convert population data to numeric
df_selected[latest_year] = pd.to_numeric(df_selected[latest_year], errors='coerce')

# Get the top 10 most populated countries
df_top10 = df_selected.sort_values(by=latest_year, ascending=False).head(10)

# Display the data to verify
df_top10
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Create a bar chart
plt.bar(df_top10['Country Name'], df_top10[latest_year], color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title(f"Top 10 Most Populated Countries ({latest_year})", fontsize=14)

# Rotate country names for better visibility
plt.xticks(rotation=45, ha='right', fontsize=10)

# Show grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
