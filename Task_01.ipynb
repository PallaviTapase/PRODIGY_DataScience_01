```python
import pandas as pd

# Replace 'your_filename.csv' with the actual downloaded file name
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv", skiprows=4)  # Skipping first 4 rows (header info)

# Display first few rows
df.head()

```

```python

# Get the latest year column
latest_year = df.columns[-1]

# Select only Country and Population columns
df_selected = df[['Country Name', latest_year]].dropna()

# Convert population values to numeric
df_selected[latest_year] = df_selected[latest_year].astype(float)

# Get the top 10 most populated countries
df_top10 = df_selected.sort_values(by=latest_year, ascending=False).head(10)

# Show the top 10 records
df_top10
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10, 5))

# Create a bar chart
plt.bar(df_top10['Country Name'], df_top10[latest_year], color='skyblue')

# Add labels and title
plt.xlabel("Country")
plt.ylabel("Population")
plt.title(f"Top 10 Most Populated Countries ({latest_year})")
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
